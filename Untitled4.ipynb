```python
import os
import numpy as np
import pandas as pd
from tqdm import tqdm
```

```python
meta_df = pd.read_excel('/Users/parkjunhyeong/Desktop/박준형/02. 대내 및 대외활동/01. 대내활동/03. Biz&AI 랩/02. 소스/01. 데이터/01. 메타 데이터/Russell_1000_combined_v0.4.xlsx', sheet_name = 'Sheet3')
news_df = pd.read_csv('/Users/parkjunhyeong/Desktop/박준형/02. 대내 및 대외활동/01. 대내활동/03. Biz&AI 랩/02. 소스/01. 데이터/02. 뉴스 데이터/04. final/nytimes.csv', index_col = 0)
```

```python
meta_df.dropna(inplace= True)
```

```python
meta_df.reset_index(inplace = True)
```

```python
meta_df
```

```python
news_df
```

```python
news_df[['Ticker', 'Description', 'GICS Sector', 'Market Cap','Name','Recheck']] = None
```

```python
for i in tqdm(range(len(news_df))):
    Tickers = []
    Descriptions = []
    GICSs = []
    Markets = []
    Names = []
    Rechecks = []
    for j in range(len(meta_df)):
        if meta_df['Name'][j] in news_df['title'][i]:
            Tickers.append(meta_df['Symbol'][j])
            Descriptions.append(meta_df['Description'][j])
            GICSs.append(meta_df['GICS Sector'][j])
            Markets.append(meta_df['Market cap'][j])
            Names.append(meta_df['Name'][j])
            Rechecks.append(meta_df['재확인'][j])
        else:
            pass
    news_df['Ticker'][i] = Tickers
    news_df['Description'][i] = Descriptions
    news_df['GICS Sector'][i] = GICSs
    news_df['Market Cap'][i] = Markets
    news_df['Name'][i] = Names
    news_df['Recheck'][i] = Rechecks
```

```python
news_df['len'] = news_df['Ticker'].apply(lambda x : len(x))
```

```python
news_df['len'].value_counts()
```

```python
news_df = news_df[news_df['len'] != 0]
```

```python
news_df.reset_index(inplace = True)
news_df.to_excel('/Users/parkjunhyeong/Desktop/temp.xlsx')
```
